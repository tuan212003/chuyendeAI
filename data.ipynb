# Tiền xử lý dữ liệu khi đã thu thập được

```python
import cv2
import os
from data import create_img_to_video,create_img_to_img,processing
from mtcnn import MTCNN
```

# Video-> ảnh

```python
input_path='datavideo'
output_path='dataoutput'

# Lặp qua các tệp trong thư mục
for filename in os.listdir(input_path):
    if filename.endswith(('.mp4')):
        print(filename)
        video_path = os.path.join(input_path, filename)
        save_path = os.path.join(output_path, os.path.splitext(filename)[0])
        create_img_to_video(video_path,save_path,20)
```

# ảnh -> ảnh


```python
input_path='dataimg'
output_path='dataoutput'

# Lặp qua các tệp trong thư mục
for filename in os.listdir(input_path):
    img_path=os.path.join(input_path, filename)
    save_path = os.path.join(output_path, os.path.splitext(filename)[0])
    create_img_to_img(img_path,save_path,20)
```
